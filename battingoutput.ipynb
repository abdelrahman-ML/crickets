```
import pandas as pd
import numpy as np

```

```
df_Batting = pd.read_csv("cricket.csv") 
# Preview the first 5 lines of the loaded data 
df_Batting.head()
```

```
df_Batting.info()
```

```
df_Batting.replace(to_replace='-',value=0,inplace=True)
```

```
df_Batting.head()
```

```
df_Batting['Runs']=pd.DataFrame(df_Batting['Runs'].astype(str).astype(int))
```

```
df_Batting.head()
```

```
df_Batting.info()
```

```
df_Batting['output1']=np.where(df_Batting['Runs']<=24, 1, 0) 
df_Batting['output2']=np.where((df_Batting['Runs']>=25) & (df_Batting['Runs']<=49), 2, 0) 
df_Batting['output3']=np.where((df_Batting['Runs']>=50) & (df_Batting['Runs']<=74), 3, 0) 
df_Batting['output4']=np.where((df_Batting['Runs']>=75) & (df_Batting['Runs']<=99), 4, 0) 
df_Batting['output5']=np.where(df_Batting['Runs']>=100 , 5, 0) 
```

```
df_Batting.head()
```

```
df_Batting.info()
```

```
#df_Batting['output']=np.logical_or(df_Batting['output1'] or df_Batting['output2'])
```

```
df_Batting['output']=df_Batting.output1 | df_Batting.output2 | df_Batting.output3 |df_Batting.output4 |df_Batting.output5
```

```
df_Batting=df_Batting.drop(columns=['output1','output2','output3','output4','output5'],axis=1)
```
